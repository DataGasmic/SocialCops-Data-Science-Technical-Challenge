```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Question - 
-  Flag set of APMC/mandis and commodities with highest price fluctuation across different commodities in each relevant season, and year

# Understanding and Intutiton of the problem
-  We shall observe fluctuation in price (max_price - min_price) for a particular year and get the PAIR
-  We shall observe fluctuation in price (max_price - min_price) for a particular season(quarter) and get the PAIR

```python
mandi_data = pd.read_csv('Deseasonalized_Mandi_Prices.csv')
msp_data = pd.read_csv('Outlier_removed_msp_data.csv')
```

# For the YEAR 2014,2015,2016 -> Yearly Price fluctuations

```python
pfluc_2014 = mandi_data[mandi_data['Year']==2014]
pfluc_2014.reset_index(inplace=True,drop=True)
pfluc_2015 = mandi_data[mandi_data['Year']==2015]
pfluc_2015.reset_index(inplace=True,drop=True)
pfluc_2016 = mandi_data[mandi_data['Year']==2016]
pfluc_2016.reset_index(inplace=True,drop=True)
```

# 2014

```python
data_2014 = pfluc_2014.groupby(['APMC','commodity']).size().reset_index()
data_2014.drop(columns=[0],inplace=True)
for i,row in data_2014.iterrows():
    temp_df = pfluc_2014[(pfluc_2014['APMC']==row.APMC)&(pfluc_2014['commodity']==row.commodity)]
    if((len(temp_df.max_price.values != 0))& (len(temp_df.max_price.values !=0))):
        fluctuation = max(temp_df.max_price.values) - min(temp_df.min_price.values)
        data_2014.loc[i,'Fluctuation'] = fluctuation
    else:
        data_2014.loc[i,'Fluctuation'] = 0
data_2014 = data_2014.sort_values(by = 'Fluctuation',ascending=False)
data_2014 = data_2014[:20]    # Taking only top 20 rows of highest fluctuations
data_2014.reset_index(inplace=True,drop=True)
data_2014
```

# 2015

```python
data_2015 = pfluc_2015.groupby(['APMC','commodity']).size().reset_index()
data_2015.drop(columns=[0],inplace=True)
for i,row in data_2015.iterrows():
    temp_df = pfluc_2015[(pfluc_2015['APMC']==row.APMC)&(pfluc_2015['commodity']==row.commodity)]
    if((len(temp_df.max_price.values != 0))& (len(temp_df.max_price.values !=0))):
        fluctuation = max(temp_df.max_price.values) - min(temp_df.min_price.values)
        data_2015.loc[i,'Fluctuation'] = fluctuation
    else:
        data_2015.loc[i,'Fluctuation'] = 0
data_2015 = data_2015.sort_values(by = 'Fluctuation',ascending=False)
data_2015 = data_2015[:20]
data_2015.reset_index(inplace=True,drop=True)
data_2015
```

# 2016

```python
data_2016 = pfluc_2016.groupby(['APMC','commodity']).size().reset_index()
data_2016.drop(columns=[0],inplace=True)
for i,row in data_2016.iterrows():
    temp_df = pfluc_2016[(pfluc_2016['APMC']==row.APMC)&(pfluc_2016['commodity']==row.commodity)]
    if((len(temp_df.max_price.values != 0))& (len(temp_df.max_price.values !=0))):
        fluctuation = max(temp_df.max_price.values) - min(temp_df.min_price.values)
        data_2016.loc[i,'Fluctuation'] = fluctuation
    else:
        data_2016.loc[i,'Fluctuation'] = 0
data_2016 = data_2016.sort_values(by = 'Fluctuation',ascending=False)
data_2016 = data_2016[:20]   
data_2016.reset_index(inplace=True,drop=True)
data_2016
```

# Observations -

```python
temp = pd.merge(data_2014,data_2015, how='inner', on=['APMC', 'commodity'])
Top_Fluc_yearly = pd.merge(temp,data_2016, how='inner', on=['APMC', 'commodity'])
Top_Fluc_yearly = Top_Fluc_yearly[['APMC','commodity']]
```

# Seasonal Fluctuation
